## PairGrid and FacetGrid

```python
import matplotlib.pyplot as plt
```

```python
flowers =sns.load_dataset('iris')
```

```python
flowers.head()
```

### PairGrid

```python
graphics = sns.PairGrid(flowers)
graphics.map(plt.scatter)
```

```python
graphics = sns.PairGrid(flowers)
graphics.map_diag(sns.distplot)
```

```python
graphics = sns.PairGrid(flowers)
graphics.map_diag(sns.distplot)
graphics.map_upper(plt.scatter)
graphics.map_lower(sns.kdeplot)
```

### FacetGrid

```python
propinas = sns.load_dataset('tips')
```

```python
propinas.head(10)
```

```python
propinas['time'].unique()
```

```python
graphics2 = sns.FacetGrid(data=propinas, col='time', row='smoker')
graphics2.map(sns.distplot, 'total_bill')
```

```python
graphics2 = sns.FacetGrid(data=propinas, col='time', row='smoker')
graphics2.map(plt.scatter, 'total_bill', 'tip')
```
