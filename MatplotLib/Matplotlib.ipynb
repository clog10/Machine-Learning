# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2, 9, 10)
x
```

```python
y = x **2
y
```

```python
plt.plot(x, y, 'red')
plt.title("This is mi graph")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'g')
plt.subplot(1,2,2)
plt.plot(y, x, 'blue')
```

## OOP graphics

```python
figure = plt.figure()
graph = figure.add_axes([0.1, 0.1, 0.8, 0.8])
graph.plot(x, y, 'purple')
```

```python
fig = plt.figure()
graph1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
graph1.plot(x, y, 'r')

graph2 = fig.add_axes([0.2, 0.4, 0.3, 0.3])
graph2.plot(y, x, 'blue')
```

## Multi Graphs (rows, columns)

```python
figu,graphs = plt.subplots(nrows=2, ncols=2)

graphs[0][0].plot(x, y, 'blue')
graphs[0][1].plot(x, y, 'red')
graphs[1][0].plot(x, y, 'yellow')
graphs[1][1].plot(x, y, 'green')

graphs[0][0].set_title('Blue')
graphs[0][1].set_title('Red')
graphs[1][0].set_title('Yellow')
graphs[1][1].set_title('Green')
```

## Size

```python
f = plt.figure(figsize=(8,4))
g = f.add_axes([0.1, 0.1, 0.9, 0.9])
g.plot(x, y, 'r')
```

## 2 graphs

```python
fi = plt.figure()
gr = fi.add_axes([0.1, 0.1, 0.9, 0.9])
gr.plot(x, x**2, label="power 2")
gr.plot(x, x**3, label="power 3")
gr.legend(loc=0)
```

## Color

```python
figu = plt.figure()
gra = figu.add_axes([0,0,1,1])
gra.plot(x, y, '#A4CCEE')
```

## Line type

```python
figur = plt.figure()
grap = figur.add_axes([0,0,1,1])
grap.plot(x, y, color = '#A4CCEE', linewidth=5, alpha = 0.3, linestyle=":")
```
