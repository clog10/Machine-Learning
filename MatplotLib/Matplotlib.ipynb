# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2, 9, 10)
x
```

```python
y = x **2
y
```

```python
plt.plot(x, y, 'red')
plt.title("This is mi graph")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'g')
plt.subplot(1,2,2)
plt.plot(y, x, 'blue')
```

## OOP graphics

```python
figure = plt.figure()
graph = figure.add_axes([0.1, 0.1, 0.8, 0.8])
graph.plot(x, y, 'purple')
```

```python
fig = plt.figure()
graph1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
graph1.plot(x, y, 'r')

graph2 = fig.add_axes([0.2, 0.4, 0.3, 0.3])
graph2.plot(y, x, 'blue')
```

## Multi Graphs (rows, columns)

```python
figu,graphs = plt.subplots(nrows=2, ncols=2)

graphs[0][0].plot(x, y, 'blue')
graphs[0][1].plot(x, y, 'red')
graphs[1][0].plot(x, y, 'yellow')
graphs[1][1].plot(x, y, 'green')

graphs[0][0].set_title('Blue')
graphs[0][1].set_title('Red')
graphs[1][0].set_title('Yellow')
graphs[1][1].set_title('Green')
```
