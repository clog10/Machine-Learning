# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2, 9, 10)
x
```

```python
y = x **2
y
```

```python
plt.plot(x, y, 'red')
plt.title("This is mi graph")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'g')
plt.subplot(1,2,2)
plt.plot(y,x')
```
